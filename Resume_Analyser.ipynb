```python
from google.colab import files

uploaded = files.upload()
```

```python
from PyPDF2 import PdfReader

filename = list(uploaded.keys())[0]

reader = PdfReader(filename)

resume_text = ""

for page in reader.pages:
    text = page.extract_text()

    if text:
        resume_text += text

print(resume_text[:3000])
```

```python
words = resume_text.split()

print("Total Characters:", len(resume_text))
print("Total Words:", len(words))
```

```python
skills_database = [
    "Python",
    "SQL",
    "Excel",
    "Power BI",
    "Tableau",
    "Machine Learning",
    "Data Analysis",
    "Pandas",
    "NumPy",
    "Statistics"
]

found_skills = []

for skill in skills_database:

    if skill.lower() in resume_text.lower():
        found_skills.append(skill)

print("Skills Found:")
print(found_skills)
```

```python
with open("resume_text.txt", "w", encoding="utf-8") as file:
    file.write(resume_text)

print("Resume text saved successfully!")
```

```python
important_skills = [
    "Python",
    "SQL",
    "Excel",
    "Power BI",
    "Tableau",
    "Machine Learning",
    "Statistics",
    "Pandas",
    "NumPy",
    "Data Analysis"
]
```

```python
matched_skills = []

for skill in important_skills:
    if skill.lower() in resume_text.lower():
        matched_skills.append(skill)

ats_score = (len(matched_skills) / len(important_skills)) * 100

print(f"ATS Score: {ats_score:.0f}/100")
```

```python
missing_skills = []

for skill in important_skills:
    if skill not in matched_skills:
        missing_skills.append(skill)

print("Missing Skills:")
print(missing_skills)
```

```python
suggestions = []

if ats_score < 60:
    suggestions.append(
        "Add more technical skills relevant to Data Analytics."
    )

if len(words) < 300:
    suggestions.append(
        "Resume appears too short."
    )

if len(words) > 1000:
    suggestions.append(
        "Resume appears too long."
    )

if len(missing_skills) > 0:
    suggestions.append(
        "Consider adding missing industry-relevant skills."
    )

print("Suggestions:")

for s in suggestions:
    print("-", s)
```

```python
report = f"""
========== ATS REPORT ==========

ATS Score: {ats_score:.0f}/100

Matched Skills:
{matched_skills}

Missing Skills:
{missing_skills}

Suggestions:
{suggestions}

===============================
"""

print(report)
```

```python
with open("ats_report.txt", "w") as file:
    file.write(report)

print("ATS Report Saved!")
```
